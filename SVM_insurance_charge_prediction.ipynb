# Support Vector Machine Algorithm -  insurance_charge_prediction
## MODEL CREATION PHASE

## read the dataset

```python
import pandas as pd 
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

## convert categorical data into numerical data

```python
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

## split Input and Output

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes' ]]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

## split train and test set 


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
x_test
```

```python
y_test
```

## Standardization (preprocessing dataset)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)   # fit - finds mean , std dev       transform - calculates and applies it
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## model creation 

```python
from sklearn.svm import SVR
#  kernel type - {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} 
regressor = SVR(kernel='rbf',C=3000)
regressor.fit(x_train,y_train)
```

## test the model 

```python
y_predict = regressor.predict(x_test)
y_predict
```

## Evaluation metrics 

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predict)
print(f" THE ACCURACY OF THE MODEL : {r_score}")     #THE ACCURACY OF THE MODEL : 0.8751977042287014   - using rbf kernel, c=3000 (after standardization)
```
